## Calculate Class Imbalance

### Subtask:
Iterate through the dataset's masks to calculate the total number of 'Artery Wall' (foreground) pixels and 'Background' pixels. Then, compute the ratio of 'Artery Wall' pixels to the total image pixels to determine class imbalance.


**Reasoning**:

```python
total_foreground_pixels = 0
total_background_pixels = 0

for item in data_dicts:
    mask_path = item["label"]
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    if mask is not None:
        total_foreground_pixels += np.sum(mask > 0)
        total_background_pixels += np.sum(mask == 0)
    else:
        print(f"Warning: Could not load mask from {mask_path}")

total_pixels = total_foreground_pixels + total_background_pixels

if total_pixels > 0:
    foreground_percentage = (total_foreground_pixels / total_pixels) * 100
    background_percentage = (total_background_pixels / total_pixels) * 100

    print(f"Class Imbalance Analysis:")
    print(f"Total Artery Wall (Foreground) Pixels: {total_foreground_pixels}")
    print(f"Total Background Pixels: {total_background_pixels}")
    print(f"Total Pixels: {total_pixels}")
    print(f"Percentage of Artery Wall (Foreground) Pixels: {foreground_percentage:.2f}%")
    print(f"Percentage of Background Pixels: {background_percentage:.2f}%")
else:
    print("No masks processed or total pixels is zero.")
```

## Define Custom CLAHE Transform

### Subtask:
Define a custom MONAI-compatible transform class, `ApplyCLAHEd`, which uses OpenCV's `cv2.createCLAHE` to apply Contrast Limited Adaptive Histogram Equalization (CLAHE) to the specified image keys. This transform will ensure consistent contrast enhancement across the dataset.


**Reasoning**:
I need to define a custom MONAI-compatible transform class, `ApplyCLAHEd`, as instructed. This involves importing necessary libraries, defining the class structure, initializing the CLAHE object with configurable parameters, and implementing the `__call__` method to apply CLAHE to specified image keys within the input dictionary, handling data type conversions for OpenCV compatibility.



```python
import cv2
import numpy as np
import torch
from monai.transforms import MapTransform

class ApplyCLAHEd(MapTransform):

    def __init__(self, keys, clipLimit=2.0, tileGridSize=(8, 8), *args, **kwargs):
        super().__init__(keys, *args, **kwargs)
        self.clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)

    def __call__(self, data):
        d = dict(data)
        for key in self.keys:
            img_tensor = d[key]

            original_device = img_tensor.device
            original_dtype = img_tensor.dtype

            if img_tensor.dim() == 3:
                if img_tensor.shape[0] == 1:
                    img_np = img_tensor.squeeze(0).cpu().numpy()
                elif img_tensor.shape[0] == 3:
                    img_np = img_tensor[0].cpu().numpy()
                else:
                    raise ValueError(f"Unsupported image shape for CLAHE: {img_tensor.shape}. Expected (C, H, W) with C=1 or C=3 (assuming grayscale duplication).")
            elif img_tensor.dim() == 2:
                img_np = img_tensor.cpu().numpy()
            else:
                raise ValueError(f"Unsupported image shape for CLAHE: {img_tensor.shape}. Expected (C, H, W) with C=1/C=3 or (H, W) for grayscale.")

            max_val = np.max(img_np)
            if max_val > 1.0 + np.finfo(float).eps:
                img_uint8 = img_np.astype(np.uint8)
                scale_factor = 1.0
            else:
                img_uint8 = (img_np * 255).astype(np.uint8)
                scale_factor = 255.0

            clahe_img_uint8 = self.clahe.apply(img_uint8)

            if scale_factor > 1.0:
                clahe_img_np = clahe_img_uint8.astype(float) / scale_factor
            else:
                clahe_img_np = clahe_img_uint8.astype(original_dtype)

            if img_tensor.dim() == 3:
                clahe_img_tensor = torch.from_numpy(clahe_img_np).unsqueeze(0).to(original_device).to(original_dtype)
            else:
                clahe_img_tensor = torch.from_numpy(clahe_img_np).to(original_device).to(original_dtype)

            d[key] = clahe_img_tensor
        return d

print("ApplyCLAHEd custom transform class defined.")
```

**Reasoning**:
The next step is to construct the `train_transforms` and `val_transforms` pipelines as specified in the main task. These pipelines will incorporate the newly defined `ApplyCLAHEd` transform along with other MONAI transforms like `LoadImaged`, `EnsureChannelFirstd`, `Resized`, `ScaleIntensityd`, `ToTensord`, and for training, `RandRotatedd`, `RandFlipd`, and `RandGaussianNoised`.



```python
from monai.transforms import ( LoadImaged, EnsureChannelFirstd, Resized, ScaleIntensityd, RandRotateD, RandFlipD, RandGaussianNoiseD, ToTensord, Compose )
from monai.data import Dataset, DataLoader

train_transforms = Compose([
    LoadImaged(keys=["image", "label"]),
    EnsureChannelFirstd(keys=["image", "label"]),
    Resized(keys=["image", "label"], spatial_size=(224, 224), mode=("bilinear", "nearest")),
    ScaleIntensityd(keys=["image"]),
    ApplyCLAHEd(keys=["image"], clipLimit=2.0, tileGridSize=(8, 8)),
    RandRotateD(keys=["image", "label"], range_x=(-15., 15.), prob=0.5, keep_size=True, mode=("bilinear", "nearest")),
    RandFlipD(keys=["image", "label"], spatial_axis=0, prob=0.5),
    RandFlipD(keys=["image", "label"], spatial_axis=1, prob=0.5),
    RandGaussianNoiseD(keys=["image"], prob=0.2, mean=0.0, std=0.1),
    ToTensord(keys=["image", "label"])
])

val_transforms = Compose([
    LoadImaged(keys=["image", "label"]),
    EnsureChannelFirstd(keys=["image", "label"]),
    Resized(keys=["image", "label"], spatial_size=(224, 224), mode=("bilinear", "nearest")),
    ScaleIntensityd(keys=["image"]),
    ApplyCLAHEd(keys=["image"], clipLimit=2.0, tileGridSize=(8, 8)),
    ToTensord(keys=["image", "label"])
])

train_ds = Dataset(data=train_dicts, transform=train_transforms)
train_loader = DataLoader(train_ds, batch_size=8, shuffle=True, num_workers=2)

val_ds = Dataset(data=val_dicts, transform=val_transforms)
val_loader = DataLoader(val_ds, batch_size=1, shuffle=False, num_workers=2)

print("Train and Validation DataLoaders with CLAHE and augmentations ready.")
```

```python
import time
from tqdm import tqdm

num_epochs = 15
best_metric_cnn = -1
best_metric_vit = -1

history = {"train_loss_cnn": [], "train_loss_vit": [], "val_dice_cnn": [], "val_dice_vit": []}

patience = 5
epochs_no_improve_cnn = 0
epochs_no_improve_vit = 0
stop_training_cnn = False
stop_training_vit = False

print(f"Starting Duel: CNN vs. USF-MAE for {num_epochs} Epochs...")

for epoch in range(num_epochs):
    if stop_training_cnn and stop_training_vit:
        print("Both models stopped due to early stopping.")
        break

    if not stop_training_cnn:
        model.train()
    if not stop_training_vit:
        model_innovation.train()

    epoch_loss_cnn = 0
    epoch_loss_vit = 0
    step = 0

    for batch_data in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        step += 1
        inputs, labels = batch_data["image"].to(device), batch_data["label"].to(device)

        if not stop_training_cnn:
            optimizer_unet.zero_grad()
            outputs_cnn = model(inputs)
            loss_cnn = criterion(outputs_cnn, labels)
            loss_cnn.backward()
            optimizer_unet.step()
            epoch_loss_cnn += loss_cnn.item()

        if not stop_training_vit:
            optimizer_usfmae.zero_grad()
            vit_inputs = inputs.repeat(1, 3, 1, 1)
            outputs_vit = model_innovation(vit_inputs)
            loss_vit = criterion(outputs_vit, labels)
            loss_vit.backward()
            optimizer_usfmae.step()
            epoch_loss_vit += loss_vit.item()

    if not stop_training_cnn:
        epoch_loss_cnn /= step
        history["train_loss_cnn"].append(epoch_loss_cnn)
    if not stop_training_vit:
        epoch_loss_vit /= step
        history["train_loss_vit"].append(epoch_loss_vit)

    if not stop_training_cnn:
        model.eval()
    if not stop_training_vit:
        model_innovation.eval()

    with torch.no_grad():
        for val_data in val_loader:
            val_inputs, val_labels = val_data["image"].to(device), val_data["label"].to(device)

            # CNN Prediction + FIX: Apply Sigmoid
            val_outputs_cnn = model(val_inputs)
            preds_cnn = (torch.sigmoid(val_outputs_cnn) > 0.5).float()
            dice_metric_unet(y_pred=preds_cnn, y=val_labels)

            # ViT Prediction + FIX: Apply Sigmoid
            vit_val_inputs = val_inputs.repeat(1, 3, 1, 1)
            val_outputs_vit = model_innovation(vit_val_inputs)
            preds_vit = (torch.sigmoid(val_outputs_vit) > 0.5).float()
            dice_metric_usfmae(y_pred=preds_vit, y=val_labels)

            if val_data["image"].shape[0] == 1: # Only print for batch size 1 for clarity
                print(f"\n[Check] CNN Max Logit: {val_outputs_cnn.max():.2f} | ViT Max Logit: {val_outputs_vit.max():.2f}")

        metric_cnn = -1 # Initialize with a value that won't trigger update
        if not stop_training_cnn:
            metric_cnn = dice_metric_unet.aggregate().item()
            dice_metric_unet.reset()
            history["val_dice_cnn"].append(metric_cnn)

        metric_vit = -1 # Initialize with a value that won't trigger update
        if not stop_training_vit:
            metric_vit = dice_metric_usfmae.aggregate().item()
            dice_metric_usfmae.reset()
            history["val_dice_vit"].append(metric_vit)

        print(f"\n--- Epoch {epoch+1} Results ---")

        if not stop_training_cnn:
            if metric_cnn > best_metric_cnn:
                best_metric_cnn = metric_cnn
                torch.save(model.state_dict(), MODEL_DIR / "best_cnn_model.pth")
                epochs_no_improve_cnn = 0
            else:
                epochs_no_improve_cnn += 1
                print(f"CNN F1 did not improve for {epochs_no_improve_cnn} epochs.")

            print(f"CNN  -> Loss: {epoch_loss_cnn:.4f}, F1-Score: {metric_cnn:.4f}")
            scheduler_unet.step(metric_cnn)

        if not stop_training_vit:
            if metric_vit > best_metric_vit:
                best_metric_vit = metric_vit
                torch.save(model_innovation.state_dict(), MODEL_DIR / "best_vit_usfmae_model.pth")
                epochs_no_improve_vit = 0
            else:
                epochs_no_improve_vit += 1
                print(f"ViT  F1 did not improve for {epochs_no_improve_vit} epochs.")
            print(f"ViT  -> Loss: {epoch_loss_vit:.4f}, F1-Score: {metric_vit:.4f}")
            scheduler_usfmae.step(metric_vit)

        if epochs_no_improve_cnn >= patience:
            stop_training_cnn = True
            print("Early stopping triggered for CNN model.")
        if epochs_no_improve_vit >= patience:
            stop_training_vit = True
            print("Early stopping triggered for ViT model.")

print(f"\nDuel Finished!")
print(f"Best CNN F1: {best_metric_cnn:.4f} | Best USF-MAE F1: {best_metric_vit:.4f}")
```
